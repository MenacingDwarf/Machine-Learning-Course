```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data.csv",sep=';')
```

```python
data.head()
```

```python
data.daily = data.milk / [monthrange(int(d[-4:]), int(d[3:5]))[1] for d in data.month]
plt.figure(figsize(15,7))
data.daily.plot()
pylab.show()
```

```python
data.daily_diff1 = data.daily - data.daily.shift(1)
data.daily_diff1 = data.daily_diff1[1:]
plt.figure(figsize(15,7))
data.daily_diff1.plot()
pylab.show()
```

```python
data.daily_diff2 = data.daily - data.daily.shift(12)
data.daily_diff2 = data.daily_diff2[12:]
plt.figure(figsize(15,7))
data.daily_diff2.plot()
pylab.show()
```

```python
data.daily_diff12 = data.daily_diff1 - data.daily_diff1.shift(12)
data.daily_diff12 = data.daily_diff12[12:]
plt.figure(figsize(15,7))
data.daily_diff12.plot()
pylab.show()
```

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(data.daily_diff12.values.squeeze(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(data.daily_diff12.values.squeeze(), lags=50)
```
