# ДЗ1 - прогнозирование временного ряда

Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

Придерживайтесь стандартного алгоритма построения прогноза:

1. Визуальный анализ ряда
2. Стабилизация дисперсии (при необходимости)
3. Выбор порядка дифференцирования
4. Выбор начальных приближений для p, q, P, Q
5. Обучение и сравнение моделей-кандидатов, выбор победителя
6. Анализ остатков построенной модели, при необходимости — её модификация: если остатки получаются смещёнными, прогноз нужно скорректировать на константу; если остатки нестационарны или автокоррелированы, можно попробовать расширить область, в которой подбираются значения p, q, P, Q.
7. Прогнозирование с помощью финальной модели.

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
salary = pd.read_csv('WAG_C_M.csv', sep=';', header=0,
                     dayfirst=True, index_col = 0, parse_dates=['month'])
salary.columns = ['salary']
salary.head()
```

---
---
---
## 1. Визуальный анализ ряда
---

```python
salary_plot = plt.plot(salary)
x_label = plt.xlabel('Year')
y_label = plt.ylabel('Salary')
pylab.show()
```

Визуально отмечаем ярко выраженный повышающийся тренд, а также годовую сезонность.

Выполним STL-декомпозицию ряда, чтобы убедиться:

```python
plt.figure(figsize(10,7))
sm.tsa.seasonal_decompose(salary.salary).plot()
plt.show()
```

#### Критерий Дики-Фуллера
Проверяем ряд на стационарность при помощи криетрия Дики-Фуллера.

---
Н0 - ряд нестационарен

Н1 - ряд стационарен

---


```python
print ('Критерий Дики-Фуллера: p = %f' % sm.tsa.stattools.adfuller(salary.salary)[1])
```

p-value получился очень большой, а значит гипотеза о нестационарности ряда не отвергается. Наш временной ряд нестационарен.

---
---
---
## 2. Стабилизация дисперсии (при необходимости)
---

#### Стабилизация дисперсии методом Бокса-Кокса

```python
salary['salary_box'], lmbda = stats.boxcox(salary.salary)
```

Проверим, стал ли ряд стационарным:

```python
print ('Преобразование Бокса-Кокса: %f' % lmbda)
print ('Критерий Дики-Фуллера: р = %f' %sm.tsa.stattools.adfuller(salary.salary_box)[1])
```

Так как р = 0.696899 , то гипотеза о нестационарности не отвергается, значит, наш временной ряд нестационарен.

---
---
---
## 3. Выбор порядка дифференцирования
---

Посмотрим на наш временной ряд после стабилизации дисперсии:

```python
salary.salary_box.plot()
plt.ylabel('Transformed salaries')
plt.xlabel('Year')
```

Можем заметить, что в нем всё еще присутствуют ярко выраженные тренд и сезонность.

Следовательно, сперва мы проведем сезонное диффериенцирование, а затем простое.

#### Сезонное диффериенцирование

```python
salary['salary_box_diff'] = salary.salary_box - salary.salary_box.shift(12)
```

Проверим, хватит ли одного сезонного диффериенцирования:

```python
print ('Криетрий Дики-Фуллера: р = %f' % sm.tsa.stattools.adfuller(salary.salary_box_diff.dropna())[1])

plt.figure(figsize(10,7))
sm.tsa.seasonal_decompose(salary.salary_box_diff.dropna()).plot()
plt.show()
```

Судя по критерию Дики-Фуллера, гипотеза о нестационарности ряда отвергается (р = 0.014697), то есть, ряд стационарен.

Однако, если мы посмотрим на STL-декомпозицию, то увидим, что у нас по прежнему остался тренд и сезонность.

Попробуем применить простое диффериенцирование.

#### Простое диффериенцирование

```python
salary['salary_box_diff2'] = salary.salary_box_diff - salary.salary_box_diff.shift(1)
```

```python
print ('Криетрий Дики-Фуллера: р = %f' % sm.tsa.stattools.adfuller(salary.salary_box_diff2.dropna())[1])

plt.figure(figsize(10,7))
sm.tsa.seasonal_decompose(salary.salary_box_diff2.dropna()).plot()
plt.show()
```

По криетрию Дики-Фуллера гипотеза о нестационарности с уверенностью отвергается.

По графикам всё еще видны изменения тренда, однако, уже менее выраженно. И остатки, в среднем, равны нулю.

# 4. Выбор начальных приближений для p, q, P, Q
---

Для того, чтобы выбрать начальные приближения для p, q, P, Q, неоюходимо проанализировать графики автокорреляции и частичной автокорреляциию

```python
plt.figure(figsize(15, 8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(salary.salary_box_diff2.dropna().values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(salary.salary_box_diff2.dropna().values.squeeze(), lags=48, ax=ax)
pylab.show()
```

Все значения берем меньше сезонности.
- Q*S - последний сезонный лаг   (acf)  (Q = лаг / S)
- q - последний несезонный лаг (acf)
- P*S - последний сезонный лаг   (pacf) (Р = лаг / S)
- р - последний несезонный лаг (pacf)

- Q=0
- q=5 (надо брать 10, но мой компьютер виснет на обучении)
- P*S=48  => P=4
- p=5 (надо брать 11, но компьютер не может посчитать)

Переберем все значения p, P, q, Q от нуля до их приближения.

```python
ps = range(0, 2) #надо брать range(0, 12), но мой компьютер виснет на обучении
qs = range(0, 2) #надо брать range(0, 11), но мой компьютер виснет на обучении
Ps = range(0, 5)
Qs = range(0, 1)
d = 1
D = 1
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

---
---
---
# 5. Обучение и сравнение моделей-кандидатов, выбор победителя
---

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(salary.salary_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
print (best_model.summary())
```

# 6. Анализ остатков построенной модели, при необходимости — её модификация: 
---
#### - если остатки получаются смещёнными, прогноз нужно скорректировать на константу; 
#### - если остатки нестационарны или автокоррелированы, можно попробовать расширить область, в которой подбираются значения p, q, P, Q.
---

#### Визуальный анализ остатков

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента), стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается коррелограммой).

Посмотрим, насколько хорошо модель описывает данные:

```python
salary['model'] = invboxcox(best_model.fittedvalues, lmbda)

plt.figure(figsize(15,7))
salary.salary.plot()
salary.model[13:].plot(color='r')
plt.ylabel('Wine sales')
pylab.show()
```

# 7. Прогнозирование с помощью финальной модели
---

- Прогноз на 2 года вперед
- Прогноз делается при помощи функции predict.
- Применить обратное преобразование к преобразованию Бокса-Кокса (invboxcox) (так как подобранная модель не значет, что искходный ряд был преобразован).

```python
salary2 = salary[['salary']]
# Задаем список дат, на которые будем делать прогноз:
date_list = [datetime.datetime.strptime('2016-08-01', '%Y-%m-%d') + relativedelta(months=x) 
             for x in range(0,24)] # 24 месяцев = 2 года 
# Создаем DataFrame с индексами - новыми датами и колонкой продаж:
future = pd.DataFrame(index=date_list, columns= salary2.columns)
# Соединяем старые данные с новыми в один датафрейм:
salary2 = pd.concat([salary2, future])
```

```python
print (len(salary2), len(salary))
```

```python
# Делаем предсказание на лучшей модели, заносим полученные предсказания в новую колонку:
salary2['forecast'] = invboxcox(best_model.predict(start=283, end=308), lmbda)

plt.figure(figsize(15,7))
salary2.salary.plot()
salary2.forecast.plot(color='r')
plt.ylabel('Wine sales')
pylab.show()
```
